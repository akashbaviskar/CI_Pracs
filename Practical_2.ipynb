```python
# ==========================================
# Clonal Selection Algorithm using Python
# (Jupyter Notebook Version)
# ==========================================

# Import Required Library
import numpy as np


# ------------------------------------------
# Objective Function
# f(x) = x^2
# Minimum value occurs at x = 0
# ------------------------------------------
def objective_function(x):
    return x ** 2


# ------------------------------------------
# Initialize Population
# ------------------------------------------
def initialize_population(size, lower_bound, upper_bound):
    return np.random.uniform(lower_bound, upper_bound, size)


# ------------------------------------------
# Clone Antibodies
# ------------------------------------------
def clone(antibodies, num_clones):
    return np.repeat(antibodies, num_clones)


# ------------------------------------------
# Hypermutation
# ------------------------------------------
def hypermutate(clones, mutation_rate):

    # Generate random mutation noise
    noise = np.random.normal(0, mutation_rate, clones.shape)

    # Return mutated clones
    return clones + noise


# ------------------------------------------
# Select Best Antibodies
# ------------------------------------------
def select_best(population, num_best):

    # Calculate fitness values
    fitness = np.array([objective_function(x) for x in population])

    # Sort according to minimum fitness
    sorted_indices = np.argsort(fitness)

    # Return best solutions
    return population[sorted_indices[:num_best]]


# ------------------------------------------
# Clonal Selection Algorithm
# ------------------------------------------
def clonal_selection_algorithm(
        pop_size=10,
        generations=20,
        clone_factor=5,
        mutation_rate=0.1,
        lower_bound=-10,
        upper_bound=10):

    # Initialize population
    population = initialize_population(
        pop_size,
        lower_bound,
        upper_bound
    )

    # Generation loop
    for gen in range(generations):

        # Select best half population
        best = select_best(population, pop_size // 2)

        # Clone selected antibodies
        clones = clone(best, clone_factor)

        # Apply hypermutation
        mutated_clones = hypermutate(clones, mutation_rate)

        # Select best individuals again
        population = select_best(mutated_clones, pop_size)

        # Find best solution
        best_solution = population[
            np.argmin([objective_function(x) for x in population])
        ]

        # Display generation result
        print(
            "Generation",
            gen + 1,
            ": Best Solution =",
            round(best_solution, 5),
            ", Fitness =",
            round(objective_function(best_solution), 5)
        )

    return best_solution


# ------------------------------------------
# Run the Algorithm
# ------------------------------------------
best = clonal_selection_algorithm()

print("\nFinal Best Solution:", round(best, 5))
```
